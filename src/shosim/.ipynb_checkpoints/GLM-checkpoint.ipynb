```python
def dif_mat(size,order):
    D1 = np.identity(size) - np.vstack((np.zeros((1,size)),np.identity(size)[:-1,:]))
    D = np.identity(size)
    k = order
    while k > 0:
        D = D @ D1
        k -= 1
    return D[order:,:]
```

```python
'''
Converts basis spline coefficients into coefficients in the polynomial basis
Params
    Theta: c_a,c_b,c_E array of coefficients for a basis spline
Returns
    Coefs: c_a-3, c_b-3,c_E-3, 4, 4, 4 dimensional array. 
           Coefs[i,j,k,q,r,s] is the coeficient on the a**q b**r E**s term in the space right after knot a_k[i+3], b_k[j+3], and E_k[k+3]
'''
def BSpline2Poly(Theta):
    D_a = a_k[1] - a_k[0]
    D_b = b_k[1] - b_k[0]
    D_E = E_k[1] - E_k[0]
    Coefs = np.zeros((Theta.shape[0] - 3,Theta.shape[1] - 3,Theta.shape[2] - 3,4,4,4))
    BSplinePieces_a = np.zeros((Theta.shape[0],4,4))
    BSplinePieces_b = np.zeros((Theta.shape[1],4,4))
    BSplinePieces_E = np.zeros((Theta.shape[2],4,4))
    '''
    BSplinePieces_?[i,j,k] holds the coefficient on the x**k term of the j-th piece of the i-th basis spline
    '''
    for i in range(Theta.shape[0]):
        for j in range(4):
            BSplinePieces_a[i,j,:] = BSplinePiece(j,a_k[0],D_a,i)
    for i in range(Theta.shape[1]):
        for j in range(4):
            BSplinePieces_b[i,j,:] = BSplinePiece(j,b_k[0],D_b,i)
    for i in range(Theta.shape[2]):
        for j in range(4):
            BSplinePieces_E[i,j,:] = BSplinePiece(j,E_k[0],D_E,i)
    for q in range(4):
        for r in range(4):
            for s in range(4):
                for l in range(4):
                    for m in range(4):
                        for n in range(4):
                            Coefs[:,:,:,q,r,s] += Theta[l:Coefs.shape[0]+l,m:Coefs.shape[1]+m,n:Coefs.shape[2]+n] \
                                * np.tile(BSplinePieces_a[l:Coefs.shape[0]+l,3-l,q].reshape((Coefs.shape[0],1,1)),(1,Coefs.shape[1],Coefs.shape[2])) \
                                * np.tile(BSplinePieces_b[m:Coefs.shape[1]+m,3-m,r].reshape((1,Coefs.shape[1],1)),(Coefs.shape[0],1,Coefs.shape[2])) \
                                * np.tile(BSplinePieces_E[n:Coefs.shape[2]+n,3-n,s].reshape((1,1,Coefs.shape[2])),(Coefs.shape[0],Coefs.shape[1],1))

    return Coefs
'''
constructs the coefficients for a 3rd order BSpline with evenly space knots in the polynomial basis
Params
    piece_num: which of the four sections of the BSpline basis element (integer 0 (leftmost) to 3 (rightmost))
    x_0: leftmost knot position
    D: distance between knots
    n: index of basis spline (0 corresponds to basis element originating from x_0)
Returns
    coefs: coefs[i] is the coefficient on the x**i term in the specified section of the basis element
'''
def BSplinePiece(piece_num,x_0,D,n):
    match piece_num:
        case 0:
            return np.array([-D**3*n**3 - 3*D**2*n**2*x_0 - 3*D*n*x_0**2 - x_0**3,3*D**2*n**2 + 6*D*n*x_0 + 3*x_0**2,-3*D*n - 3*x_0,1])/(6*D**3)
        case 1:
            return np.array([3*D**3*n**3 + 12*D**3*n**2 + 12*D**3*n + 4*D**3 + 9*D**2*n**2*x_0 + 24*D**2*n*x_0 + 12*D**2*x_0 + 9*D*n*x_0**2 + 12*D*x_0**2 + 3*x_0**3,
                            -9*D**2*n**2 - 24*D**2*n - 12*D**2 - 18*D*n*x_0 - 24*D*x_0 - 9*x_0**2,
                            9*D*n + 12*D + 9*x_0,-3])/(6*D**3)
        case 2:
            return np.array([-3*D**3*n**3 - 24*D**3*n**2 - 60*D**3*n - 44*D**3 - 9*D**2*n**2*x_0 - 48*D**2*n*x_0 - 60*D**2*x_0 - 9*D*n*x_0**2 - 24*D*x_0**2 - 3*x_0**3,
                             9*D**2*n**2 + 48*D**2*n + 60*D**2 + 18*D*n*x_0 + 48*D*x_0 + 9*x_0**2,
                             -9*D*n - 24*D - 9*x_0,3])/(6*D**3)
        case 3:
            return np.array([D**3*n**3 + 12*D**3*n**2 + 48*D**3*n + 64*D**3 + 3*D**2*n**2*x_0 + 24*D**2*n*x_0 + 48*D**2*x_0 + 3*D*n*x_0**2 + 12*D*x_0**2 + x_0**3,
                            -3*D**2*n**2 - 24*D**2*n - 48*D**2 - 6*D*n*x_0 - 24*D*x_0 - 3*x_0**2,
                            3*D*n + 12*D + 3*x_0,-1])/(6*D**3)
```

```python
'''
constructs the coefficients for a 3rd order BSpline with evenly space knots in the polynomial basis
Params
    piece_num: which of the four sections of the BSpline basis element (integer 0 (leftmost) to 3 (rightmost))
    x_0: leftmost knot position
    D: distance between knots
    n: index of basis spline (0 corresponds to basis element originating from x_0)
Returns
    coefs: coefs[i] is the coefficient on the x**i term in the specified section of the basis element
'''
def BSplinePiece(piece_num,x_0,D,n):
    match piece_num:
        case 0:
            return np.array([-D**3*n**3 - 3*D**2*n**2*x_0 - 3*D*n*x_0**2 - x_0**3,3*D**2*n**2 + 6*D*n*x_0 + 3*x_0**2,-3*D*n - 3*x_0,1])/(6*D**3)
        case 1:
            return np.array([3*D**3*n**3 + 12*D**3*n**2 + 12*D**3*n + 4*D**3 + 9*D**2*n**2*x_0 + 24*D**2*n*x_0 + 12*D**2*x_0 + 9*D*n*x_0**2 + 12*D*x_0**2 + 3*x_0**3,
                            -9*D**2*n**2 - 24*D**2*n - 12*D**2 - 18*D*n*x_0 - 24*D*x_0 - 9*x_0**2,
                            9*D*n + 12*D + 9*x_0,-3])/(6*D**3)
        case 2:
            return np.array([-3*D**3*n**3 - 24*D**3*n**2 - 60*D**3*n - 44*D**3 - 9*D**2*n**2*x_0 - 48*D**2*n*x_0 - 60*D**2*x_0 - 9*D*n*x_0**2 - 24*D*x_0**2 - 3*x_0**3,
                             9*D**2*n**2 + 48*D**2*n + 60*D**2 + 18*D*n*x_0 + 48*D*x_0 + 9*x_0**2,
                             -9*D*n - 24*D - 9*x_0,3])/(6*D**3)
        case 3:
            return np.array([D**3*n**3 + 12*D**3*n**2 + 48*D**3*n + 64*D**3 + 3*D**2*n**2*x_0 + 24*D**2*n*x_0 + 48*D**2*x_0 + 3*D*n*x_0**2 + 12*D*x_0**2 + x_0**3,
                            -3*D**2*n**2 - 24*D**2*n - 48*D**2 - 6*D*n*x_0 - 24*D*x_0 - 3*x_0**2,
                            3*D*n + 12*D + 3*x_0,-1])/(6*D**3)
```

```python
'''
Given a 6d array of coefficients, evaluate the spline at the given array of points
Params:
    Coefs: Coefs[i,j,k,q,r,s] is the coefficient on the a**q b**r E**s term in the i-th a region, j-th b region, and k-th E region
    a,b,E: input parameter values
Returns:
    Z: Result of evaluating BSpline at the provided a,b,E values
'''
def Eval_from_Coefs(a,b,E,Coefs):

    a_i = np.searchsorted(a_k[3:-3], a, side='right')
    b_i = np.searchsorted(b_k[3:-3], b, side='right')
    E_i = np.searchsorted(E_k[3:-3], E, side='right')
    
    a_i -= (a_i > Coefs.shape[0]) # so that things don't break at the upper boundaries
    b_i -= (b_i > Coefs.shape[1])
    E_i -= (E_i > Coefs.shape[2])

    Z = 0
    for l in range(4):
        for m in range(4):
            for n in range(4):
                Z += Coefs[a_i-1,b_i-1,E_i-1,l,m,n] * a**l * b**m * E**n
                
    return Z
```

```python
def Eval_from_Coefs_f(a,b,E,Coefs):
    a_i = np.searchsorted(a_k[3:-3], a, side='right')
    b_i = np.searchsorted(b_k[3:-3], b, side='right')
    E_i = np.searchsorted(E_k[3:-3], E, side='right')
    
    a_i -= (a_i > Coefs.shape[0]) # so that things don't break at the upper boundaries
    b_i -= (b_i > Coefs.shape[1])
    E_i -= (E_i > Coefs.shape[2])

    a_array = np.tile(np.stack([np.ones(np.shape(a)),a,a**2,a**3],axis=-1).reshape(*np.shape(a),4,1,1),[1,4,4])
    b_array = np.tile(np.stack([np.ones(np.shape(b)),b,b**2,b**3],axis=-1).reshape(*np.shape(b),1,4,1),[4,1,4])
    E_array = np.tile(np.stack([np.ones(np.shape(E)),E,E**2,E**3],axis=-1).reshape(*np.shape(E),1,1,4),[4,4,1])
    Z = np.sum(Coefs[a_i-1,b_i-1,E_i-1,:,:,:] * a_array * b_array * E_array,axis=(-3,-2,-1))
    return Z
```

```python
num_quad_nodes = 7
nodes_1d, weights_1d = np.polynomial.legendre.leggauss(num_quad_nodes)
weights = np.tile(weights_1d,(num_quad_nodes,1)) * np.tile(weights_1d,(num_quad_nodes,1)).T
'''
Randomly select regions of the sample space (defined by the knots)
Params:
    CoefsE: Coefficients of the probability distribution at a particular energy
    num_samples: Number of samples to take
Returns:
    a_region: array of length = num_samples. a_region[i] is the index of the region along a-axis for the ith sample
    b_region: array of length = num_samples. b_region[i] is the index of the region along b-axis for the ith sample
'''
def get_indices(CoefsE,num_samples):
    ### makes use of quadrature nodes defined in earlier cell
    nodes_ = nodes_1d.reshape(1,1,*np.ones(CoefsE.ndim-4,dtype=int),num_quad_nodes,1)
    nodesT_ = nodes_1d.reshape(1,1,*np.ones(CoefsE.ndim-4,dtype=int),1,num_quad_nodes)

    l_a = a_k[3:-4].reshape(-1,1,*np.ones(CoefsE.ndim-4,dtype=int),1,1)
    h_a = a_k[4:-3].reshape(-1,1,*np.ones(CoefsE.ndim-4,dtype=int),1,1)
    l_b = b_k[3:-4].reshape(1,-1,*np.ones(CoefsE.ndim-4,dtype=int),1,1)
    h_b = b_k[4:-3].reshape(1,-1,*np.ones(CoefsE.ndim-4,dtype=int),1,1)

    a = nodes_*(h_a-l_a)/2 + (l_a+h_a)/2
    b = nodesT_*(h_b-l_b)/2 + (l_b+h_b)/2
    a_array = np.tile(np.stack([np.ones(np.shape(a)),a,a**2,a**3],axis=-1).reshape(*np.shape(a),4,1),[1,4])
    b_array = np.tile(np.stack([np.ones(np.shape(b)),b,b**2,b**3],axis=-1).reshape(*np.shape(b),1,4),[4,1])
    
    Z = (CoefsE.reshape(*CoefsE.shape[:-2],1,1,4,4) * a_array * b_array).sum(axis=(-2,-1))
    integrated_grid = np.sum(np.exp(Z) * weights.reshape(
                                1,1,*np.ones(CoefsE.ndim-4,dtype=int),num_quad_nodes,num_quad_nodes),
                            axis=(-2,-1)) * (a_k[1] - a_k[0])*(b_k[1] - b_k[0])/4
    ranges =  np.concatenate((np.zeros((1,*integrated_grid.shape[2:])),
                              integrated_grid.reshape(-1,*integrated_grid.shape[2:]).cumsum(axis=0)),axis=0)
    
    clump_search = lambda row: np.searchsorted(row[num_samples:],row[:num_samples])
    indices = np.apply_along_axis(clump_search,0,
                                  np.concatenate((np.random.rand(num_samples,*integrated_grid.shape[2:])*ranges[-1],
                                                  ranges),axis=0)) - 1
    a_region = indices // CoefsE.shape[1]
    b_region = indices % CoefsE.shape[1]
    return a_region, b_region
    
'''
Uses quadrature nodes (specified by num_quad_nodes, nodes, and weights) to integrate the spline in the specified
rectangular bounds. Spline is specified by the precalculated coefficients at a certain energy, within a precalculated
a and b region in Coefs_abE. 
Params:
    l_a, h_a: lower and upper bounds for a integration. Can be specified as a 1d array of values but must be of shape (num_sumples,1,1)
    l_b, h_b: lower and upper bounds for b integration. Can be specified as a 1d array of values but must be of shape (num_sumples,1,1)
    Coefs_abE: Coefficients at a particular Energy, and region for a and b. Coefs_abE[sample number][a**l term][b**l term]
Returns:
    Integral over specified region
'''
### makes use of quadrature nodes defined in earlier cell
def integrate(l_a,h_a,l_b,h_b,Coefs_abE):
    nodes_ = nodes_1d.reshape(1,num_quad_nodes,1)
    nodesT_ = nodes_1d.reshape(1,1,num_quad_nodes)

    a = (nodes_*(h_a-l_a)/2 + (l_a+h_a)/2)
    b = (nodesT_*(h_b-l_b)/2 + (l_b+h_b)/2)

    a_array = np.tile(np.stack([np.ones(np.shape(a)),a,a**2,a**3],axis=-1).reshape(*np.shape(a),4,1),(1,4))
    b_array = np.tile(np.stack([np.ones(np.shape(b)),b,b**2,b**3],axis=-1).reshape(*np.shape(b),1,4),(4,1))
    
    Z = np.sum(Coefs_abE * a_array * b_array,axis=(-2,-1))
    
    return np.sum(np.exp(Z) * weights.reshape(1,num_quad_nodes,num_quad_nodes), axis=(-2,-1),keepdims=True) * (h_a-l_a)*(h_b-l_b)/4
    
'''
k
'''
def sample(Coefs,E,num_samples,sample_depth=10,binning_offset=True): 
    Eshape = np.shape(E)
    Esize = np.size(E)
    E_i = np.searchsorted(E_k[3:-3],E,side='right')
    E_i -= (E_i > Coefs.shape[2])
    E = np.reshape(E,(1,1,*np.shape(E),1,1))
    CoefsE = (Coefs[:,:,E_i-1,:,:,:] * np.stack([np.ones(np.shape(E)),E,E**2,E**3],axis=-1)).sum(axis = -1)
    a_regions, b_regions = get_indices(CoefsE,num_samples)
    Coefs_abE = CoefsE.reshape(*CoefsE.shape[:2],-1,4,4)[a_regions.reshape(num_samples,Esize,1,1),b_regions.reshape(num_samples,Esize,1,1),
                                                         np.arange(Esize).reshape(1,Esize,1,1),np.arange(4).reshape(1,1,4,1),np.arange(4).reshape(1,1,1,4)]\
                    .reshape(num_samples,*Eshape,1,1,4,4)

    l_a = np.reshape(a_k[a_regions + 3],(num_samples,*Eshape,1,1))
    h_a = np.reshape(a_k[a_regions + 4],(num_samples,*Eshape,1,1))
    l_b = np.reshape(b_k[b_regions + 3],(num_samples,*Eshape,1,1))
    h_b = np.reshape(b_k[b_regions + 4],(num_samples,*Eshape,1,1))
    
    for division in range(sample_depth):
        m_a = (l_a + h_a)/2
        Z_left = integrate(l_a,m_a,l_b,h_b,Coefs_abE)
        Z_right = integrate(m_a,h_a,l_b,h_b,Coefs_abE)
        choose_left = np.random.rand(*Z_left.shape) < (Z_left / (Z_left + Z_right))
        l_a = np.where(choose_left,l_a,m_a)
        h_a = np.where(choose_left,m_a,h_a)

        m_b = (l_b + h_b)/2
        Z_bottom = integrate(l_a,h_a,l_b,m_b,Coefs_abE)
        Z_top = integrate(l_a,h_a,m_b,h_b,Coefs_abE)
        choose_bottom = np.random.rand(*Z_bottom.shape) < (Z_bottom / (Z_bottom + Z_top))
        l_b = np.where(choose_bottom,l_b,m_b)
        h_b = np.where(choose_bottom,m_b,h_b)
    if binning_offset:
        res_a = l_a + np.random.rand(*l_a.shape)*(a_k[1]-a_k[0])/2**sample_depth
        res_b = l_b + np.random.rand(*l_a.shape)*(b_k[1]-b_k[0])/2**sample_depth
    else:
        res_a = (l_a + h_a)/2
        res_b = (l_b + h_b)/2
    return res_a.reshape(-1,*Eshape),res_b.reshape(-1,*Eshape)
```

```python
def format_energy(num):
    # Convert number to string in exponential notation with 3 decimal places
    num_str = "{:.5e}".format(num)
    # Extracting the exponent and coefficient parts
    coefficient, exponent = num_str.split('e')
    # Removing leading '+0' sign from the exponent
    exponent2 = exponent.replace('+0', '')
    if exponent2==exponent:
        raise Exception("Invalid energy for this formating. Energy must be within the interval [1,1e10)")
    # Formatting the coefficient part to have exactly 7 characters (including the '.')
    coefficient = coefficient.ljust(7, '0')
    # Return the formatted string with 9 characters
    return f"{coefficient}E{exponent2}"
energy_strs = [format_energy(num) for num in np.logspace(1,6,51)]
energies = np.logspace(1,6,51)
log_ens = np.linspace(1,6,51)
```

```python
'''
Integrate each region of the spline at a specified energy and return a grid of the integral in region
'''
def integrate_grid(Coefs,E,num_quad_nodes=7):
    nodes_1d, weights_1d = np.polynomial.legendre.leggauss(num_quad_nodes)
    weights = np.tile(weights_1d,(num_quad_nodes,1)) * np.tile(weights_1d,(num_quad_nodes,1)).T
    nodes = np.tile(nodes_1d,(num_quad_nodes,1))
    nodes_array = np.tile(nodes.reshape(1,1,num_quad_nodes,num_quad_nodes),(Coefs.shape[0],Coefs.shape[1],1,1))
    nodesT_array = np.tile(nodes.T.reshape(1,1,num_quad_nodes,num_quad_nodes),(Coefs.shape[0],Coefs.shape[1],1,1))
    weights_array = np.tile(weights.reshape(1,1,num_quad_nodes,num_quad_nodes),(Coefs.shape[0],Coefs.shape[1],1,1))
    E_i = np.vectorize(lambda E : bisect.bisect_right(E_k[3:-3],E))(E)
    E_i -= (E_i > Coefs.shape[2])

    Z = np.zeros((*Coefs.shape[:2],num_quad_nodes,num_quad_nodes))
    '''
    Z[grid of regions a][grid of regions b][grid of nodes a][grid of nodes b]
    '''
    l_a = np.tile(a_k[3:-4].reshape(-1,1,1,1),(1,Coefs.shape[1],num_quad_nodes,num_quad_nodes))
    h_a = np.tile(a_k[4:-3].reshape(-1,1,1,1),(1,Coefs.shape[1],num_quad_nodes,num_quad_nodes))
    l_b = np.tile(b_k[3:-4].reshape(1,-1,1,1),(Coefs.shape[0],1,num_quad_nodes,num_quad_nodes))
    h_b = np.tile(b_k[4:-3].reshape(1,-1,1,1),(Coefs.shape[0],1,num_quad_nodes,num_quad_nodes))
    
    for l in range(4):
        for m in range(4):
            for n in range(4):
                Z += np.tile(Coefs[:,:,E_i-1,l,m,n].reshape((Coefs.shape[0],Coefs.shape[1],1,1)),(1,1,num_quad_nodes,num_quad_nodes)) \
                        * (nodes_array*(h_a-l_a)/2 + (l_a+h_a)/2)**l * (nodesT_array*(h_b-l_b)/2 + (l_b+h_b)/2)**m * E**n
    return np.sum(np.exp(Z) * weights * (a_k[1]-a_k[0])*(b_k[1]-b_k[0])/4,axis=(2,3))
def likelihood_test(sample,Coefs):
    norm_factors = np.array([integrate_grid(Coefs,log_en).sum() for log_en in log_ens])
    return np.log(Eval_from_Coefs(sample[:,0,:],sample[:,1,:],log_ens,Coefs) / norm_factors.reshape(1,-1)).sum()
```

```python
def eval_bspline(a,b,E,Theta):
    if Theta.shape != (c_a,c_b,c_E):
        raise ValueError('Invalid shape for Theta')
    a_i = np.vectorize(lambda a : bisect.bisect_right(a_k,a))(a)
    b_i = np.vectorize(lambda b : bisect.bisect_right(b_k,b))(b)
    E_i = np.vectorize(lambda E : bisect.bisect_right(E_k,E))(E)

    a_i -= (a_i > c_a) # so that things don't break at the upper boundaries
    b_i -= (b_i > c_b)
    E_i -= (E_i > c_E)
    
    res = 0
    a_basis_spline = np.vectorize(lambda a, index : sc.interpolate.BSpline.basis_element(a_k[index:index + orders[0]+2],extrapolate = False)(a))
    b_basis_spline = np.vectorize(lambda b, index : sc.interpolate.BSpline.basis_element(b_k[index:index + orders[1]+2],extrapolate = False)(b))
    E_basis_spline = np.vectorize(lambda E, index : sc.interpolate.BSpline.basis_element(E_k[index:index + orders[2]+2],extrapolate = False)(E))
    for i in range(-orders[0]-1,0):
        for j in range(-orders[1]-1,0):
            for k in range(-orders[2]-1,0):
                res += Theta[a_i + i,b_i + j,E_i + k] \
                * a_basis_spline(a,a_i+i) \
                * b_basis_spline(b,b_i+j) \
                * E_basis_spline(E,E_i+k)
    return res
```

```python
# Make B_i and G(B_i)
x_a,x_b,x_E = (bins_a[1:] + bins_a[:-1])/2, (bins_b[1:] + bins_b[:-1])/2, log_ens
B_a = np.zeros((n_a,c_a))
B_b = np.zeros((n_b,c_b))
B_E = np.zeros((n_E,c_E))
for i in range(n_a):
    for j in range(c_a):
        B_a[i,j] = sc.interpolate.BSpline.basis_element(a_k[j:j + orders[0]+2],extrapolate = False)(x_a[i])
        if np.isnan(B_a[i,j]) : B_a[i,j] = 0
for i in range(n_b):
    for j in range(c_b):
        B_b[i,j] = sc.interpolate.BSpline.basis_element(b_k[j:j + orders[1]+2],extrapolate = False)(x_b[i])
        if np.isnan(B_b[i,j]) : B_b[i,j] = 0
for i in range(n_E):
    for j in range(c_E):
        B_E[i,j] = sc.interpolate.BSpline.basis_element(E_k[j:j + orders[2]+2],extrapolate = False)(x_E[i])
        if np.isnan(B_E[i,j]) : B_E[i,j] = 0
GB_a = np.kron(B_a,np.ones(c_a))*np.kron(np.ones(c_a),B_a)
GB_b = np.kron(B_b,np.ones(c_b))*np.kron(np.ones(c_b),B_b)
GB_E = np.kron(B_E,np.ones(c_E))*np.kron(np.ones(c_E),B_E)

GB_a_sT = sc.sparse.csr_array(GB_a.T)
GB_b_sT = sc.sparse.csr_array(GB_b.T)
GB_E_sT = sc.sparse.csr_array(GB_E.T)

# Make difference matrix, P
P_a = sc.sparse.csr_array(dif_mat(c_a,dif_orders[0]).T @ dif_mat(c_a,dif_orders[0]))
P_b = sc.sparse.csr_array(dif_mat(c_b,dif_orders[1]).T @ dif_mat(c_b,dif_orders[1]))
P_E = sc.sparse.csr_array(dif_mat(c_E,dif_orders[2]).T @ dif_mat(c_E,dif_orders[2]))

# P_s = smoothness[0] * sc.sparse.kron(np.identity(c_E * c_b), P_a) \
#   + smoothness[1] * sc.sparse.kron(sc.sparse.kron(np.identity(c_E),P_b),np.identity(c_a)) \
#   + smoothness[2] * sc.sparse.kron(P_E, np.identity(c_b * c_a))
P_s = smoothness[0] * sc.sparse.kron(P_a,np.identity(c_b * c_E)) \
  + smoothness[1] * sc.sparse.kron(sc.sparse.kron(np.identity(c_a),P_b),np.identity(c_E)) \
  + smoothness[2] * sc.sparse.kron(np.identity(c_a * c_b),P_E)
```

```python
# initial theta values
num_iters = 100
perform_likelihood_test = True
# theta = np.random.default_rng(6).random(c_a*c_b*c_E)
theta = np.random.default_rng().random(c_a*c_b*c_E)

num_samples = 4301
test_sample = np.zeros([num_samples,2,51])
for index, energy_str in enumerate(energy_strs):
    samp = Dat[energy_str].sample(n = num_samples,random_state = 2)
    test_sample[:,0,index] = ga(samp.gammaA)
    test_sample[:,1,index] = gb(samp.gammaB)
for i in range(num_iters):
    print(f'Running iteration {i}...')
    # thetas[:,i] = theta
    _ = np.moveaxis((B_a @ theta.reshape(c_a,c_b*c_E)).reshape(n_a,c_b,c_E),0,-1)
    _ = np.moveaxis((B_b @ _.reshape(c_b,c_E*n_a)).reshape(n_b,c_E,n_a),0,-1)
    Eta = np.moveaxis((B_E @ _.reshape(c_E,n_a*n_b)).reshape(n_E,n_a,n_b),0,-1)
    
    W = np.exp(Eta)
    Z = Eta + np.exp(-Eta)*(Y - np.exp(Eta))
    
    ## Make the updated (B.T * W_delta * z)
    WZ = Eta*W + Y - np.exp(Eta)
    _ = np.moveaxis((B_a.T @ WZ.reshape(n_a,n_b*n_E)).reshape(c_a,n_b,n_E),0,-1)
    _ = np.moveaxis((B_b.T @ _.reshape(n_b,n_E*c_a)).reshape(c_b,n_E,c_a),0,-1)
    BWz = np.moveaxis((B_E.T @ _.reshape(n_E,c_a*c_b)).reshape(c_E,c_a,c_b),0,-1).reshape(-1)
    W_s = sc.sparse.csc_array(W.reshape((n_a,n_b*n_E)))
    
    GW  = (GB_a_sT @ W_s).tocoo()
    H1_s = sc.sparse.csc_array((GW.data,(GW.coords[1]//n_E, c_a**2 * (GW.coords[1] % n_E) + GW.coords[0])), shape=(n_b,n_E*c_a**2))
    
    GH1 = (GB_b_sT @ H1_s).tocoo()
    H2_s = sc.sparse.csc_array((GH1.data,(GH1.coords[1]//c_a**2, c_b**2 * (GH1.coords[1] % c_a**2) + GH1.coords[0])), shape=(n_E,c_a**2 * c_b**2))
    
    GH2 = (GB_E_sT @ H2_s).tocoo()
    H3_s = sc.sparse.coo_array((GH2.data,(GH2.coords[1]//c_b**2, c_E**2 * (GH2.coords[1] % c_b**2) + GH2.coords[0])), shape=(c_a**2,c_b**2 * c_E**2))
    
    # Reshape into (B.T * W * B) matrix form
    l,m = H3_s.coords
    p = (l//c_a) % c_a
    pp = l % c_a
    q = (m//(c_b*c_E**2)) % c_b
    qp = (m//c_E**2) % c_b
    r = (m//c_E) % c_E
    rp = m % c_E
    BWB_s = sc.sparse.csr_array((H3_s.data,(p*c_b*c_E + q*c_E + r,pp*c_b*c_E + qp*c_E + rp)),shape=(c_a*c_b*c_E,c_a*c_b*c_E))
    print('    Running least squares...')
    res = sc.sparse.linalg.lsqr(BWB_s+P_s,BWz)
    theta = res[0]
    print(f'        Least squares R^2: {res[3]:5.3e}')
    print(f'        Least squares loops run: {res[2]:}')
    if perform_likelihood_test:
        print('    Performing likelihood test...')
        Coefs_sofar = BSpline2Poly(theta.reshape((c_a,c_b,c_E)))
        ll_sofar = likelihood_test(test_sample,Coefs_sofar)
        print(f'        Log-likelihood: {ll_sofar:5.3e}')
        
        
```

```python
'''''''''''
Fine tuning, run once likelihood from approximate method has leveled out
'''''''''''
num_iters = 140
ll_best = ll_sofar
theta_best = theta_untuned
theta = theta_untuned
for i in range(num_iters):
    print(f'Running iteration {i}...')
    # thetas[:,i] = theta
    _ = np.moveaxis((B_a @ theta.reshape(c_a,c_b*c_E)).reshape(n_a,c_b,c_E),0,-1)
    _ = np.moveaxis((B_b @ _.reshape(c_b,c_E*n_a)).reshape(n_b,c_E,n_a),0,-1)
    Eta = np.moveaxis((B_E @ _.reshape(c_E,n_a*n_b)).reshape(n_E,n_a,n_b),0,-1)
    
    W = np.exp(Eta)
    Z = Eta + np.exp(-Eta)*(Y - np.exp(Eta))
    
    ## Make the updated (B.T * W_delta * z)
    WZ = Eta*W + Y - np.exp(Eta)
    _ = np.moveaxis((B_a.T @ WZ.reshape(n_a,n_b*n_E)).reshape(c_a,n_b,n_E),0,-1)
    _ = np.moveaxis((B_b.T @ _.reshape(n_b,n_E*c_a)).reshape(c_b,n_E,c_a),0,-1)
    BWz = np.moveaxis((B_E.T @ _.reshape(n_E,c_a*c_b)).reshape(c_E,c_a,c_b),0,-1).reshape(-1)
    W_s = sc.sparse.csc_array(W.reshape((n_a,n_b*n_E)))
    
    GW  = (GB_a_sT @ W_s).tocoo()
    H1_s = sc.sparse.csc_array((GW.data,(GW.coords[1]//n_E, c_a**2 * (GW.coords[1] % n_E) + GW.coords[0])), shape=(n_b,n_E*c_a**2))
    
    GH1 = (GB_b_sT @ H1_s).tocoo()
    H2_s = sc.sparse.csc_array((GH1.data,(GH1.coords[1]//c_a**2, c_b**2 * (GH1.coords[1] % c_a**2) + GH1.coords[0])), shape=(n_E,c_a**2 * c_b**2))
    
    GH2 = (GB_E_sT @ H2_s).tocoo()
    H3_s = sc.sparse.coo_array((GH2.data,(GH2.coords[1]//c_b**2, c_E**2 * (GH2.coords[1] % c_b**2) + GH2.coords[0])), shape=(c_a**2,c_b**2 * c_E**2))
    
    # Reshape into (B.T * W * B) matrix form
    l,m = H3_s.coords
    p = (l//c_a) % c_a
    pp = l % c_a
    q = (m//(c_b*c_E**2)) % c_b
    qp = (m//c_E**2) % c_b
    r = (m//c_E) % c_E
    rp = m % c_E
    BWB_s = sc.sparse.csr_array((H3_s.data,(p*c_b*c_E + q*c_E + r,pp*c_b*c_E + qp*c_E + rp)),shape=(c_a*c_b*c_E,c_a*c_b*c_E))
    print('    Computing  matrix inverse...')
    BWBpPinv = sc.linalg.inv((BWB_s + P_s).todense())
    theta = BWBpPinv @ BWz
    if perform_likelihood_test:
        print('    Performing likelihood test...')
        Coefs_sofar = BSpline2Poly(theta.reshape((c_a,c_b,c_E)))
        ll_sofar = likelihood_test(test_sample,Coefs_sofar)
        if ll_sofar > ll_best:
            ll_best = ll_sofar
            theta_best = theta.copy()
        print(f'        Log-likelihood: {ll_sofar:5.3e}')
        print(f'        AIC = {ll_sofar - 2*np.trace(BWBpPinv @ BWB_s):5.3e}')
        print(f'        Effective dim = {np.trace(BWBpPinv @ BWB_s):5.3e}')
        
```
